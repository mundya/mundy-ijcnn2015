```python
import nengo
import nengo_spinnaker
import nengo.spa as spa
import numpy
import matplotlib.pyplot as plt
from nengo.utils.distributions import Uniform

def time_input(t):
    if t < 0.1:
        return "DOG"
    elif t < 0.2:
        return "CAT"
    elif t < 0.3:
        return "RAT"
    else:
        return "COW"

model = nengo.Network()
config = nengo_spinnaker.Config()

with model:
    vocab = spa.Vocabulary(16)   
    
    state = nengo.Ensemble(500, 16, label='state')
    Q = nengo.networks.EnsembleArray(64, 4, label='Q')

    
    input_node = nengo.Node(lambda t: vocab.parse(time_input(t)).v)
    config[input_node].f_of_t = True
    
    state.vocab = vocab
    nengo.Connection(state, Q.input, transform=[vocab.parse('DOG').v,
                                                vocab.parse('CAT').v,
                                                vocab.parse('RAT').v,
                                                vocab.parse('COW').v])
    
    nengo.Connection(input_node, state)
    
    input_probe = nengo.Probe(state)
    utility_probe = nengo.Probe(Q.output)
    
    bg = nengo.networks.BasalGanglia(4)
    nengo.Connection(Q.output, bg.input)
    
    bg_probe = nengo.Probe(bg.output)
    

    R = nengo.networks.EnsembleArray(64, 4, label='R',
        encoders=[[1]]*64, intercepts=Uniform(0.2, 1))
    nengo.Connection(bg.output, R.input)
    output = nengo.Probe(R.output) 
    
    bias = nengo.Node([1], label='bias')
    nengo.Connection(bias, R.input, transform=numpy.ones((4, 1)), synapse=None)

    nengo.Connection(R.output, R.input, transform=(numpy.eye(4)-1), synapse=0.008)
   
    sim = nengo_spinnaker.Simulator(model, config = config)
    sim.run(0.4)
    
    fig, axis = plt.subplots(3)
    
    axis[0].plot(sim.trange(), spa.similarity(sim.data, input_probe))
    axis[0].legend(input_probe.target.vocab.keys, fontsize='x-small')
    axis[0].set_ylabel('Input')
    
    axis[1].plot(sim.trange(), sim.data[utility_probe])
    axis[1].legend(input_probe.target.vocab.keys, fontsize='x-small')
    axis[1].set_ylabel('Utility')
    
    axis[2].plot(sim.trange(), sim.data[output])
    axis[2].legend(input_probe.target.vocab.keys, fontsize='x-small')
    axis[2].set_ylabel("Basal ganglia selection")
    
    plt.show()
```
